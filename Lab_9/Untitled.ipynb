```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
KEYWORDS = ['python', 'парсинг']
```

```python
# res.text
```

```python
req = requests.get('https://habr.com/ru/all/')
soup = BeautifulSoup(req.text, 'html.parser')
# soup
```

```python
import pandas as pd
habrDF = pd.DataFrame(columns = ['дата', 'заголовок', 'ссылка'])
index = 0
```

```python
news_blocks = soup.find_all('article', class_='post')
news_blocks
```

```python
# for post in news_blocks:
#     post_id = post.parent.attrs.get('id')
#     if not post_id:
#         continue
#     post_id = int(post_id.split('_')[-1])
#     hubs = post.find_all('a', class_='hub-link')
#     for hub in hubs:
#             hub_lower = hub.text.lower()
#             if any([hub_lower in desired for desired in KEYWORDS]):
#                 title_element = post.find('a', class_='post__title_link')
#                 print(title_element.text, title_element.attrs.get('href'))
#                 break
```

```python
for post in news_blocks:
    post_id = post.parent.attrs.get('id')
    if not post_id:
        continue
    post_id = int(post_id.split('_')[-1])
    hubs = post.find_all('a', class_='hub-link')
    for hub in hubs:
            hub_lower = hub.text.lower()
            if any([hub_lower in desired for desired in KEYWORDS]):
                title_element = post.find('a', class_='post__title_link')
                habrDF.loc[index, 'заголовок'] = title_element.text
                habrDF.loc[index, 'ссылка'] = title_element.attrs.get('href')
                time = post.find('span', class_='post__time')
                habrDF.loc[index, 'дата'] = time.text
                index = index + 1
                break
```
